```python
#importing SBopen class from mplsoccer to open the data
from mplsoccer import Sbopen
```

```python
# The first thing we have to do is open the data. We use a parser SBopen available in mplsoccer.
parser = Sbopen()
```

The data is provided as JSON files exported from the StatsBomb Data API, in the following structure:

1. Competition and seasons stored in **competitions.json**.


2. Matches for each competition and season, stored in **matches**.
    
    /Each folder within is named for a competition ID, each file is named for a season ID within that competition.

3. Events and lineups for each match, stored in events and lineups respectively. Each file is named for a **match ID**.

```python
#opening data using competition method
df_competition = parser.competition()
#structure of data
df_competition.info()
```

In order to find the event of the Match, we need to look at the competition_name  and season_name.


For example: competition_name: Copa America;
season_name: 2024


```python
df_competition['competition_name'].unique()
```

```python
df_competition['season_name'].unique()
```

Lets say, we are interested in the **Copa America of 2023/2024 Season**.

To find the match we are interested in, we need to know the competition_id  and season_id.

```python
df_competition[df_competition['season_id']==282]
```

```python
#opening data using match method
df_match = parser.match(competition_id=223, season_id=282)
#structure of data
df_match.info()
```

```python
df_match.head()
```
